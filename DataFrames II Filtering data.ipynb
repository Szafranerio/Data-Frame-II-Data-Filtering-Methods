Filter with multiple categories

```python
employees = pd.read_csv("employees.csv", parse_dates=["Start Date"], date_format="%m/%d/%Y")
employees["Last Login Time"] = pd.to_datetime(employees["Last Login Time"], format="%H:%M %p").dt.time
employees["Senior Management"] = employees["Senior Management"].astype(bool)
employees["Gender"] = employees["Gender"].astype("category")
employees.head()
```

- Add the `&` operator in between two Boolean **Series** to filter by multiple conditions.

```python
#Double category to find females in marketing and earn over 100K
female = data['Gender'] == 'Female'
marketing = data['Team'] == 'Marketing'
salary_over_100K = data['Salary'] > 100000
data [female & marketing & salary_over_100K]
```

The sign = | equal to "or" operator in between two Boolean **Series** to filter by *either* condition.

```python
#Employees who are either senior management OR started before January 1st, 1990


is_senior_management = employees["Senior Management"]
started_in_80s = employees["Start Date"] < "1990-01-01"

employees[is_senior_management | started_in_80s]
```

```python
#Combining both AND/OR filtering
is_robert = employees['First Name'] == 'Robert'
is_in_client = employees['Team'] == 'Client Services'
start_date = employees ['Start Date'] > "2016-06-01"
```

```python
employees[(is_robert & is_in_client) | start_date]
```

The isin method

```python
#Thismethod allows to filter set with the interesting values for the user. For example from the column "Team"
targeted_team = employees['Team'].isin(['Legal', 'Sales','Product'])
employees[targeted_team]
```

The isnull and notnull Methods

The `isnull` method returns True for `NaN` values in a **Series**. NaN for missing values

```python
employees[employees['Team'].isnull()]
```

The `notnull` method returns True for present values in a **Series**. WIll show set WITHOUT MISSING VALUES!!!!!!!!!!!!!!!!!!!!!!!

```python
employees[employees['Team'].notnull()]
```

The duplicated method

```python
employees['First Name'].duplicated()
```

```python
#This is showing how many duplicated values are!!!!! In that data set there were around 1000 rows and only 9 of the have UNIQUE names.
#For example Dougles can be several times as well as Thomas

employees[employees['First Name'].duplicated(keep=False)]
```

```python
#   ~   Changes the pahres to oposit. This will shows those 9 unique values
employees[~employees['First Name'].duplicated(keep=False)]
```
